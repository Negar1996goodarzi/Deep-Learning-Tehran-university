## Perceptron with Three Inputs

Task:

For the input vector x = [x1, x2, x3] = [0, 1, 1], the desired output is -1.
Use a learning rate (Î·) of 0.3.
Update the weights (W1, W2, W3) and the bias for three successive steps (epochs/phases) using the perceptron learning rule.
Note: Use the step activation function described above (threshold at 0).

```python
import numpy as np

w1_init = 0.2
w2_init = 0.7
w3_init = 0.9
bias_init = -0.7

x1 = 0
x2 = 1
x3 = 1

t = -1
alpha = 0.3
epoch = 5

w1 = np.zeros(epoch)
w2 = np.zeros(epoch)
w3 = np.zeros(epoch)
B = np.zeros(epoch)
net = np.zeros(epoch)
f = np.zeros(epoch)

w1[0] = w1_init
w2[0] = w2_init
w3[0] = w3_init
B[0] = bias_init

for i in range(epoch-1):

  net[i] = (w1[i] * x1 ) + (w2[i] * x2) + (w3[i] * x3) + B[i]
  
  # Activation function
  if net[i] > 0:
    f[i] = 1
  elif net[i] ==0:
     f[i] = 0
  else:
     f[i] = -1


  error = f[i] - t  

  if(error==0):
   print(f"Converged at iteration {i}")
   break

  else:
     w1[i + 1] = w1[i] + (alpha * t * x1)
     w2[i + 1] = w2[i] + (alpha * t * x2)
     w3[i + 1] = w3[i] + (alpha * t * x3)
     B[i + 1] = B[i] + (alpha * t)
```

```python
# Final output
print("Final weights:")
print("w1:", w1)
print("w2:", w2)
print("w3:", w3)
print("Bias:", B)
```
